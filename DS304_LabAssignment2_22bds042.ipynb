```python
# Distribution of mpg variable
# Create a histogram of the mpg variable

plt.hist(df['mpg'], bins=10, edgecolor='black')
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.title('Distribution of mpg in mtcars Dataset')
plt.show()
```

```python
# Counts for cars for each cylinder type (cyl)

cyl_counts=df['cyl'].value_counts()
print(cyl_counts)

# Plot the frequency distribution
colors = ['red', 'green', 'blue']
cyl_counts.plot(kind='bar',color=colors)
plt.xlabel('Type of cylinders')
plt.ylabel('Count')
plt.title('Frequency Distribution of Cars by Number of Gears')
plt.show()
```

```python
# Proportion of cars with automatic (am=0) & (am=1)
# Calculate the proportion of automatic cars

proportion_automatic = (df['am'] == 0).mean()
automatic = (df['am'] == 0).value_counts()


# Calculate the proportion of manual cars
proportion_manual = (df['am'] == 1).mean()
manual = (df['am'] == 1).value_counts()

print("Proportion of automatic cars:", proportion_automatic)
print("Proportion of manual cars:", proportion_manual)

print("No of automatic cars:", automatic)
print("No of manual cars:", manual)
```

```python
automatic_count=(df['am'] == 0).sum()
manual_count=(df['am'] == 1).sum()

# Create a DataFrame for plotting
data = {'Transmission': ['Automatic', 'Manual'], 'Count': [automatic_count, manual_count]}
df_1 = pd.DataFrame(data)

# Create the bar plot
colors=["blue","yellow"]
df_1.plot(x='Transmission', y='Count', kind='bar', rot=0,color=colors)
plt.title('Number of Automatic vs. Manual Cars')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Horse Power

plt.boxplot(df['hp'], vert=True)
plt.xlabel('Horsepower')
plt.title('Boxplot of horsepower in mtcars Dataset')
plt.show()
```

```python
# Distribution of cars with diffrent number of gears
# Count the frequency of each gear value

gear_counts = df['gear'].value_counts()
print(gear_counts)

# Plot the frequency distribution
colors=["orange","blue","green"]
gear_counts.plot(kind='bar',color=colors)
plt.xlabel('Number of Gears')
plt.ylabel('Count')
plt.title('Frequency Distribution of Cars by Number of Gears')
plt.show()
```

```python
# Create a grid of subplots with more axes
# Sub Plots of all the 13 diffrent shapes..

fig, axs = plt.subplots(nrows=3, ncols=5, figsize=(15, 10))

# Flatten the array of axes for easy iteration
axs = axs.flatten()

# Get the unique datasets
datasets = Df['dataset'].unique()

# Plot each dataset
for i, dataset in enumerate(datasets):
    # Filter the dataframe for the current dataset
    data = Df[Df['dataset'] == dataset]

    # Plot the data
    axs[i].scatter(data['x'], data['y'], color='blue', s=10)

    # Set the title
    axs[i].set_title(dataset)

    # Set axis limits for consistency
    axs[i].set_xlim(0, 100)
    axs[i].set_ylim(0, 100)

# Hide any remaining empty subplots
for i in range(len(datasets), len(axs)):
    axs[i].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Summary Statistics


import math

# Get the unique datasets
datasets = Df['dataset'].unique()

# Calculate the number of rows and columns needed
n_datasets = len(datasets)
n_cols = 4
n_rows = math.ceil(n_datasets / n_cols)

# Create a grid of subplots
fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 12))

# Flatten the array of axes for easy iteration
axs = axs.flatten()

# Plot each dataset and calculate summary statistics
for i, dataset in enumerate(datasets):
    # Filter the dataframe for the current dataset
    data = Df[Df['dataset'] == dataset]

    # Calculate summary statistics
    mean_x = data['x'].mean()
    mean_y = data['y'].mean()
    var_x = data['x'].var()
    var_y = data['y'].var()
    corr = data['x'].corr(data['y'])

    # Plot the data
    axs[i].scatter(data['x'], data['y'], color='blue', s=10)

    # Set the title with summary statistics
    axs[i].set_title(f"{dataset}\n"
                     f"Mean: ({mean_x:.2f}, {mean_y:.2f})\n"
                     f"Variance: ({var_x:.2f}, {var_y:.2f})\n"
                     f"Correlation: {corr:.2f}")

    # Set axis limits for consistency
    axs[i].set_xlim(0, 100)
    axs[i].set_ylim(0, 100)

# Hide any remaining empty subplots
for i in range(len(datasets), len(axs)):
    axs[i].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
